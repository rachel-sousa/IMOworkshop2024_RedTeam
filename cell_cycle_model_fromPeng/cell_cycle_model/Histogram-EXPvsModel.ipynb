```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Data for tumor survival rates
data = {
    "Concentration": [0.00, 0.10, 0.50, 1.00],
    "Model": [100.00, 92.40, 64.01, 63.99],
    "UWB/B1": [100.00, 98.73, 69.43, 61.46],
    "OVCAR3": [100.00, 94.87, 71.97, 51.70],
    "FUOV1": [100.00, 102.29, 62.36, 30.27],
    "COV318": [100.00, 86.05, 37.58, 13.63],
    "OVKATE": [99.88, 91.65, 74.70, 54.30]
}

# Data for standard deviations (SD)
sd_data = {
    "Model": [0, 0, 0, 0],
    "UWB/B1": [0.55, 5.52, 1.10, 1.99],
    "OVCAR3": [6.97, 3.80, 2.94, 6.16],
    "FUOV1": [13.21, 8.91, 8.67, 4.50],
    "COV318": [13.32, 7.77, 2.25, 0.85],
    "OVKATE": [3.42, 2.35, 0.12, 3.72]
}

# Convert to DataFrames
df = pd.DataFrame(data)
sd_df = pd.DataFrame(sd_data)

# Define colors for each model
colors = ["Black", "#377eb8", "#4daf4a", "#984ea3", "#ff7f00", "#a65628"]

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.15  # width of each bar
x = np.arange(len(df["Concentration"]))  # x-axis positions for concentrations

# Plot each model as a separate bar series with error bars
for i, (model, color) in enumerate(zip(df.columns[1:], colors)):
    ax.bar(x + i * width, df[model], width, label=model, color=color, yerr=sd_df[model], edgecolor='black')

# Labeling
ax.set_xlabel("Concentration")
ax.set_ylabel("Tumor Survival Rate (%)")
ax.set_title("ATRi")
ax.set_xticks(x + width * 2.5)
ax.set_xticklabels(df["Concentration"])
ax.legend(title="Model")

plt.tight_layout()
plt.show()
```

```python
# Data for histogram
labels = ["Model", "UWB/B1", "OVCAR3", "FUOV1", "COV318", "OVKATE"]
values = [31.7289694, 15.92352, 21.02667, 26.77596, 34.80705, 36.41827]

# Standard deviation values
errors = [0, 2.92, 1.71, 5.40, 3.14, 7.10]

# Color scheme
colors = ["Black", "#377eb8", "#4daf4a", "#984ea3", "#ff7f00", "#a65628"]

# Plot histogram with specified colors and error bars
fig, ax = plt.subplots(figsize=(8, 5), dpi=300)
bars = ax.bar(labels, values, color=colors, edgecolor='black', yerr=errors, 
              capsize=6, error_kw={'elinewidth': 2, 'capthick': 2, 'capsize': 4, 'ecolor': 'black'})

# Add circle markers at the top of each bar
for bar, value in zip(bars, values):
    ax.plot(bar.get_x() + bar.get_width() / 2, value, 'o', color='white', markeredgecolor='black', markersize=6)

# Labels and styling
ax.set_xticklabels(labels, fontweight='bold')
ax.set_ylabel("% Tumor Cell Survival", fontsize=15, fontweight='bold')
ax.set_title("PARPi-ATRi", fontsize=16, fontweight='bold')

# Customize the frame appearance
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(1.2)
ax.spines['bottom'].set_linewidth(1.2)

# Customize the ticks
ax.tick_params(axis='both', which='major', width=1.2, length=6)

plt.show()
```

```python
# Data for plotting
cell_lines = ["Model", "UWB/B1", "OVCAR3", "FUOV1", "COV318", "OVKATE"]
cdi_values = [0.719274376, 0.26, 0.41, 0.57, 0.66, 0.88]

# Color mapping (darker color for synergistic, lighter for antagonistic)
colors = ["Black", "#377eb8", "#4daf4a", "#984ea3", "#ff7f00", "#a65628"]

# Plot
fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(cdi_values, cell_lines, color=colors, s=100)  # s=100 for bubble size

# Labels and styling
ax.set_xlim(0, 2)
ax.set_xlabel("CDI")
ax.set_title("Synergistic - Antagonistic")

# Synergistic to Antagonistic Arrow
ax.annotate("Synergistic", xy=(0.2, 5.5), xytext=(0, 5.5), color="red", ha="right",
            arrowprops=dict(arrowstyle="<-", color="red"))
ax.annotate("Antagonistic", xy=(1.5, 5.5), xytext=(2, 5.5), color="blue", ha="left",
            arrowprops=dict(arrowstyle="->", color="blue"))

# Add dotted line for reference at 1 (Neutral point between synergistic and antagonistic)
ax.axvline(x=1, color="black", linestyle="dotted")

plt.show()
```
