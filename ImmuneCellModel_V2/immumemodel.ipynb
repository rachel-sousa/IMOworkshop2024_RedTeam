```python
drug_con = 0.0 #micromol/L
immune_model = create_model("ImmuneModel")
# immune_model.SetParams(C0=10, M_A0=75, M_P0=75, T0=55, K=100, q_P=1e8, q_A=1e8, r_T=0.2, beta_4=0.8, r_C=0.01)
immune_model.SetParams(C0=100, M_A0=75, M_P0=75, T0=55, r_C=0.01, d_1=5e-4, d_2=5e-4,
                      r_PA=0.15, r_AP=0.1, q_P=1e8, q_A=1e8, r_T=0.05)#, K=1000, q_P=1e8, q_A=1e8)
immune_model.Simulate(treatmentScheduleList=[[0,5.2, drug_con]])
immune_model.Trim(dt=0.5)
immune_model.Plot(plotPops=True, legend=True, drugBarPosition=0.99999, ylim=1000)
```
